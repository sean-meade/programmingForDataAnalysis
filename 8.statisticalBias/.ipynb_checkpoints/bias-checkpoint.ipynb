## Location and scale

```python
x = np.linspace(-10.0, 10.0, 1000)

plt.fill(x, ss.norm.pdf(x, loc= 0.0, scale=1.0), label="$\mu = 0.0, \sigma = 1.0$", c='b', alpha=0.6, lw=3.0)
plt.fill(x, ss.norm.pdf(x, loc= 2.0, scale=1.0), label="$\mu = 2.0, \sigma = 1.0$", c='r', alpha=0.6, lw=3.0)
plt.fill(x, ss.norm.pdf(x, loc= 0.0, scale=2.0), label="$\mu = 0.0, \sigma = 2.0$", c='g', alpha=0.6, lw=3.0)
plt.fill(x, ss.norm.pdf(x, loc=-1.0, scale=0.6), label="$\mu =-1.0, \sigma = 0.6$", c='y', alpha=0.6, lw=3.0)

plt.title("Normal distribution for different $\mu$ and $\sigma$")
plt.legend();
```

#### Above are examples of normal distribution
Each of the 4 plots above are bell shaped curves. This bell shaped curve is called the normal distribution. They follow a fairly complex looking format. They represent the measurement of a variable that typically has a mean (most likely) value called the average value (this is the peak of the curve). There is a margin of error when measuring real world values that fall away from the peak as seen above.

The normal distribution tends to be a good model for lots of things you'd want to measure in the real world.

When using a sample size like in measuring the height of people in Ireland it's likely that you haven't hit the mean height in general or in subsets exactly right. 

$ \mu \:=\: mean  $ (the average value of what you're measuring)

$ \sigma \:=\: standard\: deviation $ ($ (\sigma)^2\: $ = variance, std and/or variance measure how spread out the values typically are from the mean value)

For the blue curve the $ \mu $ is 0 and so the curve is centered at 0. The $ \sigma $ is 1 so you'll notice the curve drop off quickly. It is very unlikely that you will see numbers past 3 ($ 3\sigma $) on this type of curve in either the positive or negative direction. 

** This is an example of the standard normal distribution. It is where $ \sigma $ (standard deviation) is 1 and $ \mu $ is 0.**

The red is not standard because it is centered around 2.

** These two curves go to infinity.** (not yet sure of the other 2)

The closer the sigma is to 0 the steeper the drop off.

## Probability

When asking the probability of a value within a normal distribution you must ask for a range not a single value. The probability of a single value is technically 0. 

The curve or the probability density function of the normal distribution is represented by or follow this pattern:

$$ \int_a^b \frac{1}{\sqrt{2\pi\sigma^2} } e^{ -\frac{(x-\mu)^2}{2\sigma^2} } \, dx $$

```python
x = np.linspace(-3.0, 3.0, 1000)
y = ss.norm.pdf(x, loc= 0.0, scale=1.0)

xseg = x[np.logical_and(-1.0 < x,  x < 1.4)]
yseg = y[np.logical_and(-1.0 < x,  x < 1.4)]

plt.plot(x, y, color='k', alpha=0.5)
plt.fill_between(xseg, yseg, color='b', alpha=0.5)
plt.axvline(x=-1.0, color='grey', linestyle=':')
plt.axvline(x= 1.4, color='grey', linestyle=':')

plt.text(0.2, 0.15,
         r'$P (a \leq X \leq b) = \int_a^b \frac{1}{\sqrt{2\pi\sigma^2} } e^{ -\frac{(x-\mu)^2}{2\sigma^2} } \, dx$',
         horizontalalignment='center', size=17)

plt.axhline(y= 0.0, color='black', linestyle='-')

plt.title("Probability of $x$ between $a$ and $b$");
```

integration, remember, is the area under the curve. Getting the probability of a range is getting the percentage of the area under the curve. 

```python
x = np.linspace(-5.0, 5.0, 10000)
plt.plot(x, ss.norm.pdf(x, loc=0.0, scale=1.0), 'k-', lw=1.0)

xseg = x[np.logical_and(-5.0 <= x, x <= -3.0)]
plt.fill_between(xseg, ss.norm.pdf(xseg), color='y', alpha=0.5, linewidth=0)
xseg = x[np.logical_and(-3.0 <= x, x <= -2.0)]
plt.fill_between(xseg, ss.norm.pdf(xseg), color='r', alpha=0.5, linewidth=0)
xseg = x[np.logical_and(-2.0 <= x, x <= -1.0)]
plt.fill_between(xseg, ss.norm.pdf(xseg), color='g', alpha=0.5, linewidth=0)
xseg = x[np.logical_and(-1.0 <= x, x <=  1.0)]
plt.fill_between(xseg, ss.norm.pdf(xseg), color='b', alpha=0.5, linewidth=0, label="$1 \sigma = 68.27\%$")
xseg = x[np.logical_and( 1.0 <= x, x <=  2.0)]
plt.fill_between(xseg, ss.norm.pdf(xseg), color='g', alpha=0.5, linewidth=0, label="$2 \sigma = 95.45\%$")
xseg = x[np.logical_and( 2.0 <= x, x <=  3.0)]
plt.fill_between(xseg, ss.norm.pdf(xseg), color='r', alpha=0.5, linewidth=0, label="$3 \sigma = 99.73\%$")
xseg = x[np.logical_and( 3.0 <= x, x <=  5.0)]
plt.fill_between(xseg, ss.norm.pdf(xseg), color='y', alpha=0.5, linewidth=0)

plt.title("Normal Disribution - Probability Distribution Function")
plt.legend();
```

## Sampling distribution

In this part he's going to get the mean of a sample size of 10. He will do this for 1000 samples and then plot the distribution of these 1000 means. 

```python
# setting it so it will print out to the nearest 3 decimal places.
np.set_printoptions(formatter={'float': lambda x: "{0:6.3f}".format(x)})

# sample size
sampsize = 10
# no. of sample sizes
nosamps = 1000

# gives random numbers for a standard deviation. (do it this amount of times, each time do it on a sample size of)
samp = np.random.standard_normal((nosamps, sampsize))

print(samp)
```

```python
# calculate the mean of each sample
mean = samp.sum(axis=1) / sampsize
print(mean)
```

The sampling distribution is taking the mean of a number of samples and plotting that.

```python
# Calculate the variance.
vari = (samp - mean[:, np.newaxis])**2
vari = vari.sum(axis=1) / (sampsize)

stddev = np.sqrt(vari)
mean_stddev = stddev.sum() / nosamps

print(mean_stddev)

# Without the correction.
mean_vari = vari.sum() / nosamps
print(mean_vari)

# Using Bessel's correction: https://en.wikipedia.org/wiki/Bessel%27s_correction
print(mean_vari * (sampsize / (sampsize - 1.0)))
```
