```python
import pandas as pd
```

```python
# low_memory to get rid of warning, nrows to reduce number of rows
df = pd.read_csv("https://cli.fusio.net/cli/climate_data/webdata/dly4935.csv", skiprows=23, low_memory=False, nrows=1000)
```

```python
df
```

```python
# This should be a datetime column not object
df["date"]
```

```python
pd.to_datetime(df['date'])
```

```python
# Adds column called datetime
df['datetime'] = pd.to_datetime(df['date'])
```

```python
df = df.replace(' ', 0)
```

```python
df['temp'] = pd.to_numeric(df['maxtp'])
```

```python
# create plot with maxtp
import seaborn as sns
```

```python
df
```

```python
sns.lineplot(x='datetime', y='temp', data=df)
```
